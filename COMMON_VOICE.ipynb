```python

# Path to the folder where the sampled audio files will be copied
sampled_audio_folder = '/Users/deepaksingh/Documents/DataMbit/cv-corpus/sample_audio'  # Define the new folder

# Create the folder if it doesn't exist
if not os.path.exists(sampled_audio_folder):
    os.makedirs(sampled_audio_folder)

# Copy the first 10 audio files to the new folder
for i, audio_file in enumerate(audio_files[:10]):
    source = os.path.join(clips_folder, audio_file)
    destination = os.path.join(sampled_audio_folder, audio_file)
    shutil.copyfile(source, destination)
    print(f"Copied {audio_file} to {sampled_audio_folder}")

print("10 audio files successfully copied!")
```

```python

from google.cloud import storage

# Step 1: Set up Google Cloud credentials using the correct JSON key path
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = '/Users/deepaksingh/Documents/cloudstorage-service-account-key.json'
```

```python
# Initialize the GCS client and specify your bucket name
client = storage.Client()
bucket_name = 'genui-training'  # Your GCS bucket name
bucket = client.bucket(bucket_name)
```

```python
# Define the path to the folder with audio files
audio_clips_folder = '/Users/deepaksingh/Documents/DataMbit/cv-corpus/en/clips'  
```

```python
# Define the GCS folder where you want to upload the audio files
gcs_audio_folder = 'train_dir/Audio'
```

```python
# Upload each audio file to the GCS bucket
audio_files = [f for f in os.listdir(audio_clips_folder) if f.endswith('.mp3')]  
```

```python
for audio_file in audio_files:
    # Define the full path to the local audio file
    local_file_path = os.path.join(audio_clips_folder, audio_file)
    
    # Create a blob (file object) in the GCS bucket
    blob = bucket.blob(f'{gcs_audio_folder}/{audio_file}')  

    # Upload the file
    blob.upload_from_filename(local_file_path)
    print(f"Uploaded {audio_file} to GCS bucket {bucket_name} in folder {gcs_audio_folder}")
print("All audio files uploaded successfully!")
```
